```python
import matplotlib.pyplot as plt

labels = ['Nokia', 'Samsung', 'Apple', 'Lumia']
values = [20, 30, 45, 10]
colors =['yellow', 'green', 'red', 'blue']
explode = (0.3, 0, 0, 0)

fig = plt.figure(figsize=(4.9, 4.15))
axes = fig.add_subplot()
axes.pie(values,
         labels=labels,
         colors=colors,
         autopct='%1.2f%%',
         startangle=180.0,
         explode = explode,
         shadow=True,
         radius=1.3,
         center=(0, 0)
         )
plt.show()
```
