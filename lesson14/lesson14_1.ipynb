```python
import matplotlib.pyplot as plt

# 1.製作figure
fig = plt.figure()

# 2.在1.的figure上製作一個axes
ax_1 = fig.add_subplot(1, 1, 1)

# 3.將圖的數據儲存在axes中
X = [0, 1, 2]
Y = [3, 4, 5]
ax_1.plot(X, Y)

# 4.顯示圖
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_subplot(1,1,1)
x = [0, 1, 2]
y = [3, 4, 5]
axes.plot(x,y,'bo')
plt.show()
```

```python
import matplotlib.pyplot as plt

figure = plt.figure()
axes1 = figure.add_subplot(1,3,1)
axes1.set_title("title1")
axes2 = figure.add_subplot(1,3,2)
axes2.set_title("title2")
axes3 = figure.add_subplot(1,3,3)
axes3.set_title("title3")
plt.show()
```

```python
index = ['apple', 'banana', 'orange', 'tomato', 'guava']
values = [5, 7, 3, 4, 6]

figure = plt.figure()
axes = figure.add_subplot(1,1,1)
axes.bar(index,values)
plt.show()
```
